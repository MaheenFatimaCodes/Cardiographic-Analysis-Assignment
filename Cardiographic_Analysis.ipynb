### Checking outliers by interquartile range

```python
for col in df.select_dtypes(include="number").columns:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(col, ":", len(outliers))
```

##### Outliers were identified using the Interquartile Range (IQR) method. The analysis shows that several features contain outliers, with the highest numbers found in NSP (559), FM (347), ALTV (318), DP (284), DL (125), DS (120), MLTV (81), and MSTV (80). In contrast, LB, ASTV, Width, and Tendency each contain only 10 outliers, while UC has 13. These outliers may represent natural variations in fetal cardiotocography measurements or potential data anomalies. Therefore, they should be carefully examined before deciding whether to retain, transform, or remove them, as unnecessary removal may lead to the loss of important information and negatively impact the performance of the machine learning model.

##### **Remember:** A scatter plot should generally compare **continuous numerical variables**. Before creating one, always ask: *"Are both variables continuous measurements?"* If the answer is **yes**, a scatter plot is appropriate. If one variable represents categories or labels (even if stored as numbers), choose a box plot, violin plot, or another categorical visualization instead.
